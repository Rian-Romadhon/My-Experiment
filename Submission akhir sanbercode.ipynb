```python
#menginput data hadehh
import pandas as pd
import numpy as np

df_train = pd.read_csv('train.csv')
df_train.head()

```

```python
#menganti label gaji dengan nilai 1 dan 0
obj = {'<=7jt': 0,
       '>7jt': 1}

df_train['Gaji'] = df_train['Gaji'].replace(obj)
df_train
```

```python
df_train['Kelas Pekerja'].replace('?', 'Wiraswasta', inplace=True)
```

```python
df_train['Pekerjaan'].replace('?', 'Spesialis', inplace=True)
```

```python
# mendummies kan jenis kelamin, 
df_train = pd.get_dummies(df_train, columns=['Kelas Pekerja','Pendidikan','Status Perkawinan','Pekerjaan','Jenis Kelamin'])
df_train.info()
```

```python
#drop id biar tidak mempengaruhi prediksi
df_train = df_train.drop('id', axis=1)
df_train
```

```python
# bagi menjado data training dan testing
x= df_train.drop('Gaji', axis=1)

#standarisasi data
#from sklearn.preprocessing import StandardScaler
#SS = StandardScaler()
#trainingSS = SS.fit_transform(x)
#x_train = pd.DataFrame(trainingSS, columns=x.keys())



y = df_train['Gaji']
```

```python
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=21, stratify=y)
```

```python
#from sklearn.neighbors import KNeighborsClassifier

#model = KNeighborsClassifier(n_neighbors=10, weights='distance')
#model.fit(x_train, y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=12, splitter='best', random_state=30, min_samples_split=13, min_samples_leaf=3)
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)

from sklearn.metrics import roc_auc_score
roc = roc_auc_score(y_test, y_predict)
roc
```
